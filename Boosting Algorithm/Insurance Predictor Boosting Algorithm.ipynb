```python
import pandas as pd
```

```python
dataset = pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset = pd.get_dummies(dataset,dtype= int, drop_first = True)
dataset
```

```python
dataset.columns
```

```python
independent = dataset[["age","bmi","children","sex_male","smoker_yes"]]
dependent = dataset[["charges"]]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(independent,dependent,test_size = 0.3,random_state = 0)
```

Gradient Boosting

```python
#Gradient Boosting
from sklearn.ensemble import GradientBoostingRegressor
gb_regressor = GradientBoostingRegressor(criterion = "friedman_mse",n_estimators=100)
gb_regressor.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score
y_gb_pred = gb_regressor.predict(x_test)
gb_r2_score = round(r2_score(y_test,y_gb_pred),2)
print(f"Gradient Boosting R2 Score : {gb_r2_score}")
```

```python
import xgboost as xg

xg_regressor = xg.XGBRegressor(objective='reg:squarederror', n_estimators=100, learning_rate=0.1)
xg_regressor.fit(x_train, y_train)
```

```python
y_xg_pred = xg_regressor.predict(x_test)
xg_r2_score = round(r2_score(y_test,y_xg_pred),2)
print(f"XG Boosting R2 Score : {xg_r2_score}")
```

Adaboosting

```python
#XG Boosting
from sklearn.ensemble import AdaBoostRegressor
ada_regressor = AdaBoostRegressor(random_state=0, n_estimators=100)
ada_regressor.fit(x_train,y_train)
```

```python
y_ada_pred = ada_regressor.predict(x_test)
ada_r2_score = round(r2_score(y_test,y_ada_pred),2)
print(f"Ada Boosting R2 Score : {ada_r2_score}")
```
